## Useful Oprations

### Filter & Where

```python
df
```

```python
df.groupby("Company")[["Age"]].mean()
```

```python
df.groupby("Company").filter(lambda x: x["Age"].mean() < 30)

# şirket olarak 2 ye ayırdı. Amazonun yaş ortalaması 30 dan küçük olduğu için bütün Amazon verilerini getirdi. 
# OpenAI yaş ortalaması 30 dan büyük olduğu için hiçbirini getirmedi...
```

```python
df
```

```python
df[(df["Company"] == "Amazon") & (df["Age"] < 30)]

# Hem Amazon olup hem de yaşı 30 dan küçük olanlar..
```

```python
df[df["Age"] < 30]
```

```python
df.groupby("Company").filter(lambda x: x["Age"].mean() > 30)
```

```python
df.groupby("Company")[["Salary"]].mean()
```

```python
df.groupby("Company").filter(lambda x: x["Salary"].mean() > 180)
```

```python
df[df.Age > 30]
```

```python
df.where(df.Age > 30)
```

```python
df.where(df.Age > 30, 0)   # nan yerlerin hepsine 0 koyduk..
```

```python
df[df.Age > 30]
```

```python
df.where(df.Age > 30).dropna()
```

## Transform

```python
df
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df
```

```python
df.groupby("Department")["Salary"].transform("mean")
```

```python
df["Salary_mean"] = df.groupby("Department")["Salary"].transform("mean")   
df

# yeni bir sütun oluşturacaksak transform daha mantıklıdır. Bütün valuelara değer atadığı için..
```

```python
df
```

```python
df["difference"] = df.Salary - df.Salary_mean
df
```

```python
df
```

```python
df.Salary.transform(lambda x: x * 1.2)
```

```python
df["New_salary"] = df.Salary.transform(lambda x: x * 1.2)
df
```
