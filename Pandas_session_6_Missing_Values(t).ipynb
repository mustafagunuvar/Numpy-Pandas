## Missing Value Handling Methods

## dropna() & drop()

```python
df
```

```python
df.dropna()   # içerisinde bir tane bile NaN değer olan satırları komple siler.
```

```python
df
```

```python
df.dropna(axis= 0, how= "any", inplace= False)   # any belirtilen satır içerisinde bir tane bile NaN değer varsa siler
```

```python
df.dropna(axis= 0, how= "all", inplace= False)  # eğer bir satırda bütün değerler NaN ise siler. 
```

```python
df.dropna(axis = 1, how= "any", inplace=False)   # sütun bazlı bakar. 
```

```python
df.dropna(axis = 1, how= "all", inplace=False)
```

```python
df.dropna(axis= 1, thresh= 9, inplace= False)  # en az 9 tane Notnull değer varsa dursun aksi halde sil.. 
```

```python
df.drop()   # içerisine silmek istediğimiz satır veya sütunu yazmamız lazım. 
```
